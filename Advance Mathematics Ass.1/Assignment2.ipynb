```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

```python
data = pd.read_csv('data.csv',encoding='ISO-8859-1')
data
```

```python
x = pd.to_numeric(data['no2'],errors = 'coerce').dropna().values
x
```

```python
roll_no = 102317112
```

```python
ar = 0.05 * (roll_no % 7)
br = 0.3 * ((roll_no % 5) + 1)
```

```python
ar
```

```python
br
```

```python
z = x + ar * np.sin(br*x)
z
```

```python
counts,bin_edges = np.histogram(z,bins=100,density=True)
bin_centers = (bin_edges[:-1] + bin_edges[1:])/2
```

```python
def pdf(z_val,l,mu,c):
  return c * np.exp(-l * (z_val - mu) ** 2)
```

```python
initial_guess = [0.01,np.mean(z),np.max(counts)]
params,_ = curve_fit(pdf,bin_centers,counts,p0 = initial_guess)
l_final,mu_final,c_final = params
```

```python
print(f"Lambda : {l_final:.5f}")
print(f"Mu : {mu_final:.5f}")
print(f"c : {c_final:.5f}")
```
